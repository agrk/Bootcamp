**(1)** Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')
sns.set(font = 'Arial')
sns.set_style("white")
df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
title_style = {'family': 'Arial', 'color': 'darkred', 'size': 20 }
axis_style  = {'family': 'Arial', 'color': 'darkblue', 'size': 15 }
```

```python
asi = df[df['location'].isin(['Spain', 'France', 'Germany', 'Italy']) & (df['date'] >= pd.to_datetime("2020-04-01"))]
```

```python
plt.figure(figsize=(15, 10), dpi = 100)
sns.lineplot(x = asi.date, y = asi.total_deaths, data = asi, hue = 'location',lw=2)
plt.title('Total Deaths', fontdict = title_style)
plt.xlabel('Dates', fontdict = axis_style)
plt.ylabel('Count of Deaths', fontdict = axis_style)
plt.xticks(rotation = 30, fontsize = 9)
plt.show()
```

**(3)**  Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

#### REFERENCE ####

```python
new_data=df[df['location']=='France']
```

```python
plt.figure(figsize=(16, 16), dpi = 100)

plt.subplot(2,2,1)
sns.distplot(new_data.new_deaths.dropna(), kde= False, bins=50,color = '#FF914D')
plt.title('New Deaths\n(Bins:50 / KDE = False)',fontdict = title_style)
plt.xlabel('New Deaths', fontdict = axis_style)
plt.ylabel('Count', fontdict = axis_style)

plt.subplot(2,2,2)
sns.distplot(new_data.new_deaths.dropna(), kde = False, bins=20,color = '#22B573')
plt.title('New Deaths\n(Bins:20 / KDE = False)',fontdict = title_style)
plt.xlabel('New Deaths', fontdict = axis_style)
plt.ylabel('Count', fontdict = axis_style)

plt.subplot(2,2,3)
sns.distplot(new_data.new_deaths.dropna(), kde = True, bins=50,color = '#22B573')
plt.title('New Deaths\n(Bins:50 / KDE = True)',fontdict = title_style)
plt.xlabel('New Deaths', fontdict = axis_style)
plt.ylabel('Normalized Count', fontdict = axis_style)

plt.subplot(2,2,4)
sns.distplot(new_data.new_deaths.dropna(), kde = True, bins=20,color = '#22B573')
plt.title('New Deaths\n(Bins:20 / KDE = True)',fontdict = title_style)
plt.xlabel('New Deaths', fontdict = axis_style)
plt.ylabel('Normalized Count', fontdict = axis_style)

plt.show()
```

**(4)** Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

#### REFERENCE ####

```python
two_country=df[df['location'].isin(['France', 'Germany'])]
```

```python
plt.figure(figsize=(15, 6),dpi=100)



plt.subplot(1,3,1)
sns.boxplot(x=two_country.location, y=two_country.new_deaths, data=two_country)
plt.ylabel('New Deaths', fontdict = axis_style)
plt.xlabel('Country', fontdict = axis_style)
plt.title('Box Plot',fontdict = title_style)


plt.subplot(1,3,2)
sns.violinplot(x=two_country.location, y=two_country.new_deaths, data=two_country)
plt.ylabel('New Deaths', fontdict = axis_style)
plt.xlabel('Country', fontdict = axis_style)
plt.title('Violin Plot',fontdict = title_style)


plt.subplot(1,3,3)
sns.swarmplot(x=two_country.location, y=two_country.new_deaths, data=two_country)
plt.ylabel('New Deaths', fontdict = axis_style)
plt.xlabel('Country', fontdict = axis_style)
plt.title('Box Plot',fontdict = title_style)


plt.show()
```
