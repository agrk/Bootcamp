In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

```python
df.sort_values(['location','date']).head(1)
```

#### 1) First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. ####

```python
spain=pd.DataFrame(df[df['location'] == 'Spain'][-100:])
```

```python
france=pd.DataFrame(df[df['location'] == 'France'][-100:])
```

```python
germany=pd.DataFrame(df[df['location'] == 'Germany'][-100:])
```

```python
italy=pd.DataFrame(df[df['location'] == 'Italy'][-100:])
```

#### 2) Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries ####

```python
plt.figure(figsize=(20, 5))

plt.subplot(121)
plt.plot(spain.date,spain.total_cases)
plt.plot(france.date,france.total_cases)
plt.plot(germany.date,germany.total_cases)
plt.plot(italy.date,italy.total_cases)


plt.subplot(122)
plt.plot(spain.date,spain.total_deaths)
plt.plot(france.date,france.total_deaths)
plt.plot(germany.date,germany.total_deaths)
plt.plot(italy.date,italy.total_deaths)
plt.show()
```

#### 3) Set titles, X and Y labels, change the font, font size, and font color for each subplot. ####

```python
plt.figure(figsize=(20, 5))

plt.title('GDP per Capita', 
          fontstyle = 'normal', # {'normal', 'italic', 'oblique'}
          fontsize = 20,
          fontweight = 'normal', # {a numeric value in range 0-1000, 'normal', 'regular', etc.}
          color = "darkred",
          fontfamily = 'Arial'
         )

plt.subplot(121).set_title('DATE & TOTAL CASES',fontdict={'size':20,'weight':'bold', 'color': '#FFC300','family': 'Rockwell'})
plt.subplot(121).set_xlabel('DATES',fontdict={'size':18, 'color': 'r','family': 'Franklin Gothic Medium', 'fontstyle':'italic'})
plt.subplot(121).set_ylabel('TOTAL CASES',fontdict={'size':18, 'color': 'b','family': 'Franklin Gothic Medium', 'fontstyle':'italic'})
plt.plot(spain.date,spain.total_cases)
plt.plot(france.date,france.total_cases)
plt.plot(germany.date,germany.total_cases)
plt.plot(italy.date,italy.total_cases)


plt.subplot(122).set_title('DATE & TOTAL DEATH',fontdict={'size':20,'weight':'bold', 'color': '#1ABC9C','family': 'Rockwell'})
plt.subplot(122).set_xlabel('DATES',fontdict={'size':18, 'color': 'r','family': 'Franklin Gothic Medium', 'fontstyle':'italic'})
plt.subplot(122).set_ylabel('TOTAL DEATH',fontdict={'size':18, 'color': 'b','family': 'Franklin Gothic Medium', 'fontstyle':'italic'})
plt.plot(spain.date,spain.total_deaths)
plt.plot(france.date,france.total_deaths)
plt.plot(germany.date,germany.total_deaths)
plt.plot(italy.date,italy.total_deaths)
plt.show()
```

#### 4) Change color, width, and type of lines. ####

```python
plt.figure(figsize=(25, 10))

plt.subplot(121).set_title('DATE & TOTAL CASES',fontdict={'size':15,'weight':'bold', 'color': '#FFC300'})
plt.subplot(121).set_xlabel('DATES',fontdict={'size':18, 'color': 'r'})
plt.subplot(121).set_ylabel('TOTAL CASES',fontdict={'size':18, 'color': 'r'})
plt.plot(spain.date,spain.total_cases, c = "#FF914D", marker = "+", ms = 5, linestyle = ":", lw = 4)
plt.plot(france.date,france.total_cases, c = "#C0392B", marker = "+", ms = 5, linestyle = "--", lw = 4)
plt.plot(germany.date,germany.total_cases, c = "purple", marker = "*", ms = 5, linestyle = "-", lw = 4)
plt.plot(italy.date,italy.total_cases, c = "#22B573", marker = "o", ms = 5, linestyle = "-.", lw = 4)


plt.subplot(122).set_title('DATE & TOTAL DEATH',fontdict={'size':15,'weight':'bold', 'color': '#1ABC9C'})
plt.subplot(122).set_xlabel('DATES',fontdict={'size':18, 'color': 'r'})
plt.subplot(122).set_ylabel('TOTAL DEATH',fontdict={'size':18, 'color': 'r'})
plt.plot(spain.date,spain.total_deaths, c = "#FF914D", marker = "+", ms = 5, linestyle = ":", lw = 4)
plt.plot(france.date,france.total_deaths, c = "#C0392B", marker = "+", ms = 5, linestyle = "--", lw = 4)
plt.plot(germany.date,germany.total_deaths, c = "purple", marker = "*", ms = 5, linestyle = "-", lw = 4)
plt.plot(italy.date,italy.total_deaths, c = "#22B573", marker = "o", ms = 5, linestyle = "-.", lw = 4)
plt.show()
```

#### 5) Put a legend to bottom right. ####

```python

plt.figure(figsize=(25, 10))

plt.subplot(121).set_title('DATE & TOTAL CASES',fontdict={'size':15,'weight':'bold', 'color': '#FFC300'})
plt.subplot(121).set_xlabel('DATES',fontdict={'size':18, 'color': 'r'})
plt.subplot(121).set_ylabel('TOTAL CASES',fontdict={'size':18, 'color': 'r'})
plt.plot(spain.date,spain.total_cases, c = "#FF914D", marker = "+", ms = 5, linestyle = ":", lw = 4)
plt.plot(france.date,france.total_cases, c = "#C0392B", marker = "+", ms = 5, linestyle = "--", lw = 4)
plt.plot(germany.date,germany.total_cases, c = "purple", marker = "*", ms = 5, linestyle = "-", lw = 4)
plt.plot(italy.date,italy.total_cases, c = "#22B573", marker = "o", ms = 5, linestyle = "-.", lw = 4)
plt.legend(['SPAIN', 'FRANCE', 'GERMANY', 'ITALY'],title = 'Countries', loc = 4)

plt.subplot(122).set_title('DATE & TOTAL DEATH',fontdict={'size':15,'weight':'bold', 'color': '#1ABC9C'})
plt.subplot(122).set_xlabel('DATES',fontdict={'size':18, 'color': 'r'})
plt.subplot(122).set_ylabel('TOTAL DEATH',fontdict={'size':18, 'color': 'r'})
plt.plot(spain.date,spain.total_deaths, c = "#FF914D", marker = "+", ms = 5, linestyle = ":", lw = 4)
plt.plot(france.date,france.total_deaths, c = "#C0392B", marker = "+", ms = 5, linestyle = "--", lw = 4)
plt.plot(germany.date,germany.total_deaths, c = "purple", marker = "*", ms = 5, linestyle = "-", lw = 4)
plt.plot(italy.date,italy.total_deaths, c = "#22B573", marker = "o", ms = 5, linestyle = "-.", lw = 4)
plt.legend(['SPAIN', 'FRANCE', 'GERMANY', 'ITALY'],title = 'Countries', loc = 4)

plt.show()
```
